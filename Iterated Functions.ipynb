```python
import numpy as np
import matplotlib.pyplot as plt
```

### Function 1
$$x_{n+1} = rx_n(1-x_n)$$

```python
def f(r,x):
    return r*x*(1-x)
```

```python
x_test = np.linspace(0,1)
plt.plot(x_test,f(2,x_test))
plt.show()


plt.plot(x_test,f(2,x_test))
plt.plot(x_test,x_test)
plt.savefig("nobif.png")
plt.show()
plt.plot(x_test,f(3.1,x_test))
plt.plot(x_test,x_test)
plt.savefig("bif.png")
```

```python
x_1000 = []

r = np.linspace(0,4,1000)

for r in r:
    sol = []
    sol.append(np.random.random())
    for i in range(1000):
        sol.append(f(r,sol[-1]))
        if i == 999:
            x_1000.append(sol[-1])
```

```python
r = np.linspace(0,4,1000)
plt.plot(r,x_1000,',k',alpha=.5)
plt.savefig("iterated.eps")
```

### Function 2
$$x_{n+1} = rsin(x_n)$$

```python
def f_1(r,x):
    return r*np.sin(x)
```

```python
x_test = np.linspace(0,np.pi)
plt.plot(x_test,f_1(2,x_test))
plt.show()


plt.plot(x_test,f_1(1.5,x_test))
plt.plot(x_test,x_test)
plt.show()
plt.plot(x_test,f_1(2.1,x_test))
plt.plot(x_test,x_test)
```

### Function 3
$$x_{n+1} = re^{-(x_n)^2}$$

```python
def f_2(r,x):
    return r*np.exp(-x**2)
```

```python
x_test = np.linspace(0,np.pi)
plt.plot(x_test,f_2(2,x_test))
plt.show()


plt.plot(x_test,f_2(1.5,x_test))
plt.plot(x_test,x_test)
plt.show()
plt.plot(x_test,f_2(2.1,x_test))
plt.plot(x_test,x_test)
```

```python
x_1000 = []

r = np.linspace(0,5,1000)

for r in r:
    sol = []
    sol.append(np.random.random())
    for i in range(1000):
        sol.append(f_2(r,sol[-1]))
        if i == 999:
            x_1000.append(sol[-1])
```

### Function 5
$$x_{n+1} = rx_n(1-x_n)+\epsilon r(x_{n})^2$$

```python
def f_4(r,x):
    e = 0.5
    return r*x*(1-x) + e*r*(x**2+np.sin(x**2+2))
```

```python
x_test = np.linspace(0,2)
plt.plot(x_test,f_4(2,x_test))
plt.show()


plt.plot(x_test,f_4(1.5,x_test))
plt.plot(x_test,x_test)
plt.show()
plt.plot(x_test,f_4(2.1,x_test))
plt.plot(x_test,x_test)
```

```python
x_1000 = []

r = np.linspace(0,2,1000)

for r in r:
    sol = []
    sol.append(np.random.random())
    for i in range(1000):
        sol.append(f_4(r,sol[-1]))
        if i == 999:
            x_1000.append(sol[-1])
```

```python
r = np.linspace(0,2,1000)
plt.plot(r,x_1000,',k',alpha=.5)
plt.legend()
plt.savefig("iterated_5.png")
```
